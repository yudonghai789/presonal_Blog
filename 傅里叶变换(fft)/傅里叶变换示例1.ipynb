## Fourier Transform

正变换：

$$
X(f)=\int_{-\infty}^{\infty}x(t)e^{-j2\pi ft}dt
$$

逆变换：

$$
x(t)=\int_{-\infty}^{\infty}X(f)e^{j2\pi ft}df
$$


## Discrete Fourier Transform

$$
X[k]=\sum_{n=0}^{N-1}x[n]e^{-j2\pi kn/N}
$$

$$
x[n]=\frac{1}{N}\sum_{k=0}^{N-1}X[k]e^{j2\pi kn/N}
$$


## 生成时域信号

两个正弦波叠加

```python
import numpy as np
import matplotlib.pyplot as plt

fs=500
t=np.linspace(0,1,fs)

f1=5
f2=20

signal=np.sin(2*np.pi*f1*t)+np.sin(2*np.pi*f2*t)

plt.figure(figsize=(8,4))
plt.plot(t,signal)
plt.title('Time Domain Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid()
plt.show()
```

## FFT 计算频谱

```python
X=np.fft.fft(signal)
freqs=np.fft.fftfreq(len(X),1/fs)

mask=freqs>=0

plt.figure(figsize=(8,4))
plt.stem(freqs[mask],np.abs(X[mask]))
plt.xlim(0,50)
plt.title('Frequency Spectrum')
plt.xlabel('Frequency')
plt.ylabel('Magnitude')
plt.grid()
plt.show()
```

可以看到两个峰：

- 5 Hz
- 20 Hz

这正是构造信号时使用的频率。

## 交互实验

拖动频率，观察时域与频域变化

```python
from ipywidgets import interact,FloatSlider

def explore(f1=5,f2=20):

    signal=np.sin(2*np.pi*f1*t)+np.sin(2*np.pi*f2*t)

    X=np.fft.fft(signal)
    freqs=np.fft.fftfreq(len(X),1/fs)
    mask=freqs>=0

    plt.figure(figsize=(12,4))

    plt.subplot(1,2,1)
    plt.plot(t,signal)
    plt.title(f'Time Signal f1={f1}Hz f2={f2}Hz')
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.grid()

    plt.subplot(1,2,2)
    plt.stem(freqs[mask],np.abs(X[mask]))
    plt.xlim(0,50)
    plt.title('Frequency Spectrum')
    plt.xlabel('Frequency')
    plt.ylabel('Magnitude')
    plt.grid()

    plt.show()

interact(
    explore,
    f1=FloatSlider(min=1,max=40,step=1,value=5),
    f2=FloatSlider(min=1,max=40,step=1,value=20)
)
```
