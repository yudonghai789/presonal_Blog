## Fourier Transform

$$
X(f)=\int_{-\infty}^{\infty}x(t)e^{-j2\pi ft}dt
$$

$$
x(t)=\int_{-\infty}^{\infty}X(f)e^{j2\pi ft}df
$$


## 导入库

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider, IntSlider
```

# 实验 1：傅里叶合成

通过控制频率生成时域信号

```python
fs=500
t=np.linspace(0,1,fs)

def synth(f1=5,f2=20):

    signal=np.sin(2*np.pi*f1*t)+np.sin(2*np.pi*f2*t)

    plt.figure(figsize=(8,4))
    plt.plot(t,signal)
    plt.title(f'Signal f1={f1}Hz f2={f2}Hz')
    plt.grid()
    plt.show()

interact(
    synth,
    f1=FloatSlider(min=1,max=40,step=1,value=5),
    f2=FloatSlider(min=1,max=40,step=1,value=20)
)
```

# 实验 2：FFT 频谱分析

```python
def fft_demo(f1=5,f2=20):

    signal=np.sin(2*np.pi*f1*t)+np.sin(2*np.pi*f2*t)

    X=np.fft.fft(signal)
    freqs=np.fft.fftfreq(len(X),1/fs)

    mask=freqs>=0

    plt.figure(figsize=(12,4))

    plt.subplot(1,2,1)
    plt.plot(t,signal)
    plt.title('Time Domain')
    plt.grid()

    plt.subplot(1,2,2)
    plt.stem(freqs[mask],np.abs(X[mask]))
    plt.xlim(0,50)
    plt.title('Frequency Spectrum')
    plt.grid()

    plt.show()

interact(
    fft_demo,
    f1=FloatSlider(min=1,max=40,step=1,value=5),
    f2=FloatSlider(min=1,max=40,step=1,value=20)
)
```

# 实验 3：傅里叶级数逼近方波

```python
t=np.linspace(-np.pi,np.pi,1000)

def square_series(N=1):

    y=np.zeros_like(t)

    for n in range(1,N+1):
        y+=np.sin((2*n-1)*t)/(2*n-1)

    y*=4/np.pi

    plt.figure(figsize=(8,4))
    plt.plot(t,y)
    plt.title(f'Fourier Series Approx N={N}')
    plt.grid()
    plt.show()

interact(square_series,N=IntSlider(min=1,max=100,step=1,value=1))
```

# 实验 4：频域滤波

```python
def filter_demo(cutoff=10):

    signal=np.sin(2*np.pi*5*t)+np.sin(2*np.pi*20*t)

    X=np.fft.fft(signal)
    freqs=np.fft.fftfreq(len(X),1/fs)

    X_filtered=X.copy()
    X_filtered[np.abs(freqs)>cutoff]=0

    signal_filtered=np.fft.ifft(X_filtered)

    plt.figure(figsize=(10,4))

    plt.subplot(1,2,1)
    plt.plot(t,signal)
    plt.title('Original Signal')

    plt.subplot(1,2,2)
    plt.plot(t,signal_filtered.real)
    plt.title('Filtered Signal')

    plt.show()

interact(filter_demo,cutoff=FloatSlider(min=1,max=30,step=1,value=10))
```
