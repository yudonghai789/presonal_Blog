# python虚拟环境
> 为什么需要按照python虚拟环境
  因为不同的项目需要不用的库，多个项目之间存在冲突。 不同的虚拟环境之间可以相互隔离





```python
安装 Python 虚拟环境（Virtual Environment）是开发中的“基本功”。简单来说，它能为每个项目创建一个独立、隔离的 Python 运行环境，避免不同项目之间因为库版本冲突而打架。

目前最主流、最推荐的方法是使用 Python 自带的 venv 模块。

🛠 使用 venv 创建虚拟环境（推荐）
这是最轻量、无需安装额外工具的方法。

1. 创建虚拟环境
首先，在终端（Terminal 或 PowerShell）中进入你的项目文件夹，然后运行：

Bash

# windows
python -m venv .venv

# macOS / Linux
python3 -m venv .venv
提示： .venv 是虚拟环境的名字，你可以随便起（比如 my_env），但习惯上大家用 .venv。

2. 激活虚拟环境
创建完后，你需要“进入”这个环境：

Windows (PowerShell):

PowerShell

.\.venv\Scripts\Activate.ps1
Windows (CMD):

DOS

.\.venv\Scripts\activate.bat
macOS / Linux:

Bash

source .venv/bin/activate
激活成功标志： 你的命令行提示符前面会出现 (.venv) 字样。

3. 安装库与使用
现在你可以放心安装任何库，它们只会待在这个小房间里：

Bash

pip install requests
4. 退出环境
当你忙完这个项目，想回到全局环境时：

Bash

deactivate
```
