# 在当前文件夹创建 Jupyter Notebook 运行环境

这篇博客将教你如何在你当前的工作文件夹中快速搭建一个 Jupyter Notebook 环境，并能直接运行 Python 代码。

## 1️⃣ 安装 Python 和 pip

确保你已经安装了 Python3 和 pip。你可以在终端或命令提示符中运行：

````python
python --version  # 或 python3 --version
pip --version     # 或 pip3 --version
````

如果没有安装 Python，可以去 [Python 官网](https://www.python.org/downloads/) 下载并安装。

## 2️⃣ 在当前文件夹创建虚拟环境

在你的目标文件夹打开终端，然后运行：

````python
# 在当前目录创建一个名为 'venv' 的虚拟环境
python -m venv venv
````

### 激活虚拟环境
- **Windows:**
```cmd
venv\Scripts\activate
```
- **Linux/Mac:**
```bash
source venv/bin/activate
```

激活后，你会看到终端前面多了 `(venv)`，表示虚拟环境已经生效。

## 3️⃣ 安装 Jupyter Notebook

在虚拟环境中运行：

````python
pip install notebook
````

## 4️⃣ 在当前文件夹启动 Jupyter Notebook

运行：

````python
jupyter notebook
````

这会在浏览器中打开 Jupyter Notebook 页面，默认路径就是你当前的文件夹。

你可以在这个页面中创建新的 `.ipynb` 文件并开始写代码。

## 5️⃣ 退出和关闭虚拟环境

关闭 Jupyter Notebook 页面，然后在终端运行：

````python
deactivate
````

---
✅ **总结**：
1. 安装 Python 和 pip
2. 在当前文件夹创建虚拟环境
3. 激活虚拟环境
4. 安装 Jupyter Notebook
5. 启动 Notebook 并开始编写代码
6. 使用完毕后退出虚拟环境

这样，你就有了一个干净、独立的 Jupyter Notebook 运行环境，所有依赖都在当前文件夹管理，不会影响全局 Python。
