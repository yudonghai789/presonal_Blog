# Python 虚拟环境：从混乱到有序的必经之路

作为 Python 开发者，你是否遇到过这样的困境：项目 A 需要 `Django 2.2`，而项目 B 必须在 `Django 4.0` 上运行？直接在全局环境安装库会导致版本冲突，甚至可能搞挂你的系统 Python。

**虚拟环境 (Virtual Environment)** 就是为了解决这个问题而生的“隔离舱”。

## 1. 为什么需要虚拟环境？

1. **版本隔离**：不同项目依赖不同版本的包，互不干扰。
2. **权限管理**：无需管理员（sudo）权限即可安装库。
3. **环境一致性**：方便导出依赖清单（requirements.txt），确保团队成员环境一致。



## 2. 使用内置的 `venv` 模块

从 Python 3.3 开始，标准库自带了 `venv`。这是最推荐、最轻量的方法。

### 第一步：创建环境
在你的项目根目录下运行（`myenv` 是环境名称，你可以自定义）：

```python
# Windows
!python -m venv myenv

# macOS / Linux
# !python3 -m venv myenv
```

### 第二步：激活环境
激活后，你在终端输入 `python` 将指向该虚拟环境。 

> **注意**：在 Jupyter Notebook 中，激活命令需要在外部终端执行。这里仅展示命令格式。

- **Windows**: `myenv\Scripts\activate` 
- **macOS/Linux**: `source myenv/bin/activate` 

激活成功后，你的命令行提示符前通常会出现 `(myenv)` 字样。

## 3. 管理依赖包

进入环境后，你可以自由安装所需的库。

```python
# 安装一个示例库
!pip install requests

# 生成依赖清单
!pip freeze > requirements.txt
```

## 4. 在 Jupyter 中使用虚拟环境

如果你想让这个 Notebook 运行在刚才创建的虚拟环境中，需要安装 `ipykernel`：

```python
# 在激活的环境中运行
!pip install ipykernel
!python -m ipykernel install --user --name=myenv --display-name "Python (MyEnv)"
```

安装完成后，刷新页面，在菜单栏的 **Kernel -> Change Kernel** 中选择 "Python (MyEnv)" 即可。

## 总结

| 操作 | 命令 |
| :--- | :--- |
| 创建 | `python -m venv <name>` |
| 激活 (Win) | `<name>\Scripts\activate` |
| 激活 (Unix) | `source <name>/bin/activate` |
| 退出 | `deactivate` |

养成“一个项目，一个环境”的习惯，能帮你规避 90% 的环境配置报错！
